# HW0: Wine Dataset
Below, you can find my preliminary exploration of the wine data set.

For reference, here are the abbreviated labels I used in my code.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import parallel_coordinates
from pandas.tools.plotting import radviz
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 10, 10
```

```python
df_wine = pd.read_csv('wine.csv')
print df_wine.columns.values
ls_cols = ['Alc', 'Hue', 'Phenols', 'C Intensity', 'Class']
df_viz = df_wine[ls_cols]
```

We'll be focusing on just the following fields: Alcohol, Hue, Total Phenols, and Color Intensity

I'll start by exploring singular attributes.

```python
parallel_coordinates(df_viz, 'Class', colormap='Set1')
```

In the above parallel coordinates plot, it seems as though the Alcohol and Color Intensity features offer decent separation. However, the nature of the plot means the lines are overlapped, so it's difficult to see the extent of this overlap. Let's try something different.

```python
df_viz1 = df_wine[['Alc', 'C Intensity', 'Class']]
df_viz2 = df_wine[['Hue', 'Phenols', 'Class']]
pylab.rcParams['figure.figsize'] = 10, 5
df_viz1.boxplot(by='Class')
df_viz2.boxplot(by='Class')
```

These box plots help us better understand the extent of the overlap among the classes for each of the chosen variables. It seems that no one feature will be a great identifier for the class labels. Let's try a couple visualizations that consider multiple features.

```python
pylab.rcParams['figure.figsize'] = 10, 10
radviz(df_viz, 'Class', colormap='Set1')
```

The above so-called 'RadViz' uses a basic spring tension model to visualize multivariate data. In this particular plot, we can see some fairly clear separation among the three classes.

```python
df_viz.plot(kind='scatter', x='Alc', y='Hue', c='Class', s=50, colormap='Set1', lw=0)
```
